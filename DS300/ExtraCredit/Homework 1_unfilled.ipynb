How many names are you able to recover?

#### Linking Attack


The anonymous data set only contains 2 rows with the same DOB and zip. This means up to 32,559 names could be recovered using the pii table.

```python
# Remove groups not able to link
unique_data = adult_anon.groupby(['DOB', 'Zip']).size() # get count of DOB, Zip rows
tuple_count = unique_data.reset_index(name="count") # make variable out of count value
tuple_count = tuple_count[tuple_count['count'] < 2] # filter out counts that cannot be linked with these variables
```

```python
# In this cell, write code to perform the linking attack
linkable_pii = tuple_count.merge(pii, 'left', ['DOB', 'Zip']) # add rest of pii to include names in linking attack
linkable_pii = linkable_pii.drop('count', 1) # drop count column
linked_data = linkable_pii.merge(adult_anon, 'left', ['DOB', 'Zip']) # linking attack results
linked_data.head()
```
